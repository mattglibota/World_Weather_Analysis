```python
#import dependencies
import pandas as pd, matplotlib.pyplot as plt, numpy as np

# Import linear regression from the SciPy stats module.
from scipy.stats import linregress

#import citypy
from citipy import citipy

#import datetime
from datetime import datetime

#import requests
import requests

#import API key
from config import weather_api_key
```

```python
#create set of random lat and longs
dps = 1500
lats = np.random.uniform(low=-90.000,high=90.000,size=dps)
lngs = np.random.uniform(low=-180.000,high=180.000,size=dps)
lat_lngs = list(zip(lats,lngs))
```

```python
#list for holding cities
cities = []
#identify nearest city using citipy
for coordinate in lat_lngs:
    city = citipy.nearest_city(coordinate[0],coordinate[1]).city_name
    
    #if it is unique, add to list
    if city not in cities:
        cities.append(city)
        
#print city count to confirm sufficient data set
len(cities)
```

```python
# Starting URL for Weather Map API Call.
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key
#city_url = url + "&q=Nanaimo"
#city_weather = requests.get(city_url).json()
#city_weather["rain"]
```

```python
# create the empty list
city_data = []
# print start of loggings
print("Beginning Data Retrieval     ")
print("-----------------------------")

# Create counters.
record_count = 1
set_count = 1

# loop through all the cities
for i, city in enumerate(cities):
    
    #group cities in sets of 50
    if (i % 50 == 0 and i >=50):
        set_count+=1
        record_count = 1
    
    #create endpoint URL for each city
    city_url = url + "&q=" + city
    
    #log url, record and set number and city
    print(f"Processing Record {record_count} of Set {set_count} | {city}")
    record_count +=1
    
    try:
        #parse json
        city_weather = requests.get(city_url).json()
        #parse out needed data
        city_country = city_weather["sys"]["country"]
        city_lat = city_weather["coord"]["lat"]
        city_lng = city_weather["coord"]["lon"]
        city_max_temp = city_weather["main"]["temp_max"]
        city_humidity = city_weather["main"]["humidity"]
        city_clouds = city_weather["clouds"]["all"]
        city_wind = city_weather["wind"]["speed"]

        #convert date to ISO standard
        city_date = datetime.utcfromtimestamp(city_weather["dt"]).strftime('%Y-%m-%d %H:%M:%S')
        
        try:
            #get weather descriptors
            city_weather_desc = city_weather["weather"][0]["description"]
            city_weather_type = city_weather["weather"][0]["main"]
        except:
            print("Bad Desc")
        
        #get rain precipitation with try
        try:
            city_rain = city_weather["rain"]["3h"]
        except:
            city_rain = 0
        
        #get snow precipitation with try
        try:
            city_snow = city_weather["snow"]["3h"]
        except:
            city_snow = 0
        
        #append to the empty city datas list
        city_data.append({"City": city.title(),
                    "Country": city_country,
                    "Date": city_date,
                    "Lat": city_lat,
                    "Lng": city_lng,
                    "Max Temp": city_max_temp,
                    "Humidity": city_humidity,
                    "Cloudiness": city_clouds,
                    "Wind Speed": city_wind,
                    "Current Desc": city_weather_desc,
                    "Rain Inches (last 3 hrs)": city_rain,
                    "Snow Inches (last 3 hrs)": city_snow}
              )
        
        
    except:
        print(f"City name {city} not found. Skipping...")
        pass
    
# Indicate that Data Loading is complete.
print("-----------------------------")
print("Data Retrieval Complete      ")
print("-----------------------------")
```

```python
#convert the array of dicts to a DataFrame
city_df = pd.DataFrame(city_data)
city_df.head(5)
```

```python
#create output file
output_data_file = "weather_data/WeatherPy_challenge.csv"
#export to file
city_df.to_csv(output_data_file,index_label="City_ID")
```
